```python
def prepare_outlines():
  User_Prompt = f"""
  I am writing a report for 'Cosmix wellness' company, here are the headings that i will pass on to RAG system which will give answers on those headings, so generate questions for each of these headings, include the company's name in each question.
    I. Company Overview
        A. What does the company do?
          1. Industry
              - In which industry does the company operate?
              - What are the key characteristics of this industry?
          2. Core business activities
              - What are the main products or services offered by the company?
              - What are the company's primary business activities?
          3. Target market
              - Who are the company's target customers?
              - What are the demographics and psychographics of the target market?
        B. Management and Experience
          1. Key management personnel
              - Who are the key members of the management team?
              - What are their roles and responsibilities within the company?
          2. Background and experience of management team
              - What is the educational background of the management team members?
              - What relevant experience do they have in the industry or related fields?
          3. Organizational structure
              - How is the company structured organizationally?
              - What are the main departments or divisions within the company?
    II. Product Analysis
        A. Products sold by the company
            1. Product categories
              - What are the main product categories offered by the company?
              - How are these products categorized or grouped?
            2. Key features and benefits
              - What are the key features of each product category?
              - What benefits do these products provide to customers?
            3. Pricing strategy
              - What is the company's pricing strategy for its products?
              - How does the pricing compare to competitors' products?
        B. Product Reviews
            1. Customer feedback and testimonials
              - What do customers say about the company's products?
              - Are there any notable customer testimonials or reviews?
            2. Professional reviews and ratings
              - Are there any professional reviews or ratings of the company's products?
              - What are the strengths and weaknesses highlighted in these reviews?
            3. Strengths and weaknesses of the products
              - What are the main strengths of the company's products?
              - Are there any significant weaknesses or areas for improvement?
    III. Competitive Landscape
        A. Identification of competitors
          1. Direct competitors
              - Who are the company's main direct competitors?
              - What products or services do these competitors offer?
          2. Indirect competitors
              - Are there any indirect competitors or substitutes for the company's products?
              - How do these indirect competitors impact the market?
        B. Competitor analysis
          1. Market share
              - What is the market share of the company compared to its competitors?
              - How has the market share evolved over time?
          2. Competitive advantages and disadvantages
              - What are the competitive advantages of the company's products or services?
              - What are the disadvantages or challenges faced by the company in the market?
          3. Strategies employed by competitors
              - What strategies do competitors use to attract customers?
              - How do competitors differentiate themselves in the market?
        C. Company's position in the market
          1. Unique selling proposition (USP)
              - What is the company's unique selling proposition?
              - How does the company differentiate itself from competitors?
          2. Competitive strengths and weaknesses
              - What are the company's main competitive strengths?
              - Are there any significant competitive weaknesses?
          3. Opportunities and threats
              - What opportunities exist for the company in the market?
              - Are there any threats or challenges that the company needs to address?
    IV. Conclusion
      A. Summary of key findings
          - What are the most important takeaways from the analysis?
          - How can the company leverage its strengths and address its weaknesses?
      B. Recommendations for the company
          - Based on the analysis, what recommendations can be made to the company?
          - What strategies or actions should the company consider to improve its market position?
      C. Future outlook and potential for growth
          - What is the future outlook for the company and its products?
          - Are there any potential growth opportunities or areas for expansion?

  Respond with a python parsable list, where each item has these fields
   - heading: name of the topic
   - questions: questions related to the topic
  """

  messages = [{"role": "user", "content": User_Prompt}]
  response = completion(model="openrouter/anthropic/claude-3-haiku", messages=messages)
  response_text = response['choices'][0]['message']['content']
  return response_text
```

```python
questions = prepare_outlines()
print(questions)
```

```python
def get_answers_per_question(heading, question):
  User_Prompt = f"""
  Can you provide detailed data on {question}
  """
  print(User_Prompt)
  messages = [{"role": "user", "content": User_Prompt}]
  response = completion(model="openrouter/perplexity/sonar-medium-online", messages=messages)
  return response.choices[0].message.content
```
